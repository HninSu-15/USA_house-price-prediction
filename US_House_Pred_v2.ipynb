```python
import kagglehub

path = kagglehub.dataset_download("shree1992/housedata")
print("Path:", path)
```

```python
import pandas as pd
import os

# Find CSV file automatically
for file in os.listdir(path):
    if file.endswith('.csv'):
        csv_path = os.path.join(path, file)
        print("CSV file:", csv_path)

df = pd.read_csv(csv_path)
df
```

```python
df.describe()
```

```python
df = df[df['price'] > 0]
df['price'].min()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df_clean['price'])
plt.title("Price Outliers")
plt.show()
```

```python
numeric_df = df_clean.select_dtypes(include=['int64', 'float64'])

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Feature Selection

Previous model use three features sqft_living,bathrooms and sqft_ above, the model's r2 score is just 0.35 ,means that the model explains 35% of the variability in house prices using features like sqft_living, bathrooms, sqft_above. So I will use other important features in this training.

In this training, we consider the important of categorical features,to measure the relationship between the categorical features and target , we can't directly use the standard correlation method.

```python
# Use group mean price for city
df.groupby('city')['price'].mean().sort_values(ascending=False)
```

Because the p-value is extremely small (much lower than 0.05), we can conclude that house prices differ significantly between cities. This means the `city` feature has a strong impact on house price and should be included in the predictive model (after proper encoding).

A p-value < 0.05 means the feature significantly affects the target.
Higher F-value â†’ stronger difference between group means.

```python
#ANOVA test
from scipy import stats

# Example: compare price across different cities
categories = df_clean['city'].unique()
groups = [df_clean[df_clean['city']==cat]['price'] for cat in categories]
f_val, p_val = stats.f_oneway(*groups)
print("F-value:", f_val, "p-value:", p_val)
```

```python
#ANOVA test for statezip
from scipy import stats

# Example: compare price across different cities
categories = df_clean['statezip'].unique()
groups = [df_clean[df_clean['statezip']==cat]['price'] for cat in categories]
f_val, p_val = stats.f_oneway(*groups)
print("F-value:", f_val, "p-value:", p_val)
```

New features, floors,view and bedrooms and categorical features city and statezip are considered the selected features.

```python

select_feature = ['bathrooms','sqft_living','sqft_above','floors','view','bedrooms','city','statezip']
X =df_clean[select_feature]
y= df_clean['price']

# Encode the categorical using one_hot encoding

X = pd.get_dummies(X, columns=['city', 'statezip'], drop_first=True)


# import the library to split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,
                                                test_size=0.3,
                                                random_state=42)

# print the training and testing size
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)

```

# Model Selection
Polynomial REgression is not suitable for many categorical Features because one-hot encoder creates many columns when encoding. So, use the Random Forest which can learn nonlinear relationship automatically ,good for house price prediction.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
rf = RandomForestRegressor(
    n_estimators=300,     # number of trees
    max_depth=20,       # let trees grow fully
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1             # use all CPU cores
)

rf.fit(X_train, y_train)
```

#Model Evaluation after using RandomForest Model and use new features

```python
# Test Performance between predict and Actual
# Predict the price for test data
import numpy as np
ytest_pred =rf.predict(X_test)

# Predict the price for train data
ytrain_pred = rf.predict(X_train)

df_train_result = pd.DataFrame({'Actual':np.round(y_train),'Predict_values': np.round(ytrain_pred)})
df_train_result
```

```python
df_test_result = pd.DataFrame({'Actual':np.round(y_test),'Predicted values':np.round(ytest_pred)})
df_test_result
```

```python
#import the required modules

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#Evaluate on training set
mae = mean_absolute_error(y_train,ytrain_pred)
print("Mean bsolute error: ",np.round(mae))
mse = mean_squared_error(y_train,ytrain_pred)
print("Mean squared error:",mse)
r2 = r2_score(y_train,ytrain_pred)
print("R2 score is :", r2)

#Evaluate on testing set
mae = mean_absolute_error(y_test,ytest_pred)
print("Mean bsolute error: ",np.round(mae))
mse = mean_squared_error(y_test,ytest_pred)
print("Mean squared error:",mse)
r2 = r2_score(y_test,ytest_pred)
print("R2 score is :", r2)
```
